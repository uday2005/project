```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot multiple lines
plt.plot(x, y1, label="y = x^2", color="blue", linestyle="--", marker="o")
plt.plot(x, y2, label="y = x", color="red", linestyle="-.", marker="s")

# Add labels, legend, and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

# Show the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

# Create scatter plot
plt.scatter(x, y, color="green", marker="x", s=100)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")

# Show the plot
plt.show()
```

```python
# Data
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

# Create bar chart
plt.bar(categories, values, color="purple")

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")

# Show the plot
plt.show()
```

```python
# Create horizontal bar chart
plt.barh(categories, values, color="orange")

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Bar Chart")

# Show the plot
plt.show()
```

```python
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]

# Create histogram
plt.hist(data, bins=5, color="blue", edgecolor="black")

# Add labels and title
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")

# Show the plot
plt.show()
```

```python
# Data
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 30, 20, 10]
colors = ["gold", "lightblue", "lightgreen", "pink"]
explode = (0.1, 0, 0, 0)  # Highlight the first slice

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y, color="red", linewidth=2, linestyle="--", marker="o", markersize=8)

# Add grid, labels, and title
plt.grid(True, linestyle="--", alpha=0.5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Plot")

# Show the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Create subplots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y1, color="blue")
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.plot(x, y2, color="green")
plt.title("Plot 2")

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create plot
plt.plot(x, y)
plt.title("Save Plot Example")

# Save the plot
plt.savefig("plot_example.png")

# Show the plot
plt.show()
```
